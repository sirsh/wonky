```python
#testing some bizarre CRN
rsys =  wonky.reaction.reaction_system([wonky.reaction("A", "A+A", 0.35),
                                        wonky.reaction("A", "0", 0.35),
                                        wonky.reaction("B", "B+B", 0.35),
                                        wonky.reaction("B", "0", 0.35),
                                        wonky.reaction("C", "C", 0.2),
                                        wonky.reaction("C", "0", 0.8),
                                        wonky.reaction("A+B", "C", 0.1)
                                       ])
net = wonky.crn(rsys, (9,2), ICs=[1,1,0])
net.display_trace(2)
```

```python
net = wonky.crn(rsys, (9,2), ICs=[2,2,0])
net.sample(display=True,params={"name":"../sample_data/exp/01"})
net.stats.observable(obs=1)
```

```python
import numpy as np
import pandas as pd
#turn this into a function - if species = None, stack all the species and return this transposed data frame
species_traces = pd.DataFrame(np.stack([net.stats[i][0][1] for i in range(3)])).T
species_traces.plot()
```
