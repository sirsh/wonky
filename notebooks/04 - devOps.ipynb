# 03 Reproducibility
Depending on how we define reproducibility, to reproduce an experiment i need the source control version, the parameter space meta data and the random number generator seed.

In the sample project, for every experiment run, the binary numpy files are stored in a folder along with a meta file with this info. 

In principle each experiment is given a unique identifier and can be archived systematically. The rock solid way is for every experiment to be given a GUID but that is a bit nasty in practice (for humans). 

User-given names, possibly with dates generated by the code, can be used depending on taste. A simple archive script can automate this in a consistent way with custom logic.

```python
import os
import pathlib
import shutil
def archive(epath, overwrite=False, default_archive_location="./"):
    """
    Some naive archiving 
    """
    #maybe I have an archive for all my data and I could store it's location as a system variable
    archive_location = os.environ["WonkyArchive"] if "WonkyArchive" in os.environ else default_archive_location
    P = pathlib.Path(epath)
    apath = os.path.join(archive_location, P.parts[-1])
    print("archiving to {}.zip".format(apath))
    #make the archive directory if it does not exist
    if not os.path.exists(archive_location):os.makedirs(archive_location)
    if os.path.isfile("{}.zip".format(apath)): 
        user_likes = input("File exists already - (o)verwrite or (k)eep both? ")
        if user_likes == "o":pass
        elif user_likes == "k": print("Write some code to manage file names")
        else: pass # default or do nothing      
    shutil.make_archive(apath, 'zip', epath)      
```
